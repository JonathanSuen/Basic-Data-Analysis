```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

Import the insurance.csv as pandas dataframe

```python
Insurance_data = pd.read_csv("insurance.csv")
```

Define the x & y value

```python
x = Insurance_data.loc[:, 'age'].values
y = Insurance_data.loc[:, 'charges'].values
```

Scatter plot with age on the x-axis and charges on the y-axis

```python
plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title("How age affects insurance cost")
plt.show()
```

Fit the model with Linear regression of the independent variable x against dependent variable y

```python
insurance_model = LinearRegression()
X = x.reshape(-1, 1)
insurance_model.fit(X, y)
print("The coef of the model is: ", insurance_model.coef_)
print("The intercept of the model is: ", insurance_model.intercept_)
```

Predict the value using the regression model and plot the scatter plot with the best-fit line

```python
y_pred = insurance_model.predict(X)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title("How age affects insurance cost")
plt.show()
```

Cross-validation to evaluate the model

```python
scores = cross_val_score(insurance_model, X, y, cv=10)
print("Cross-validation scores: {}".format(scores))
```

```python
mean_score = np.mean(scores)
std_score = np.std(scores)
print("Mean cross-validation score: {:.2f}".format(mean_score))
print("Standard deviation of cross-validation scores: {:.2f}".format(std_score))
print("According to the mean score it is too low than 1 which imply Linear Regression is not a good fit model within age and charges.")
```
